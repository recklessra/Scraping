```python
from time import sleep
from selenium import webdriver
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()

#options.add_argument('--headless')
options.add_argument('--incognito')

driver = webdriver.Chrome(options=options)
driver.implicitly_wait(10)

driver.get('https://atsumaru.jp/area/7/list?sagid=all')
sleep(3)

    
soup = BeautifulSoup(driver.page_source, 'lxml')
jobs = soup.select('div.bl_card2')

base_url = 'https://atsumaru.jp'

for job in jobs[:3]:
    company_info = job.select_one('span.bl_card2_ttl_text > a')
    
    company_name = company_info.text
    print(company_name)
    
    company_url = base_url + company_info.get('href')
    #print(company_url)
    
    detail_driver = webdriver.Chrome(options=options)
    detail_driver.implicitly_wait(10)
    
    detail_driver.get(company_url)
    sleep(3)
    
    detail_soup = BeautifulSoup(detail_driver.page_source, 'lxml')
    add = detail_soup.select_one('div.bl_card2_infoWrap > div:nth-of-type(3) p:first-of-type').text
    print(add)
    
    button = detail_driver.find_element_by_css_selector('a.bl_contactNav_btn__tel')
    button.click()
    sleep(3)
    
    phone_no_soup = BeautifulSoup(detail_driver.page_source, 'lxml')
    phone_no = phone_no_soup.select_one('div.telNo > i').text
    
    print(phone_no)
    
    detail_driver.quit()

print(len(jobs))

driver.quit()
#driver.close()
```
