```python
from bs4 import BeautifulSoup

html = """
<!DOCTYPE html>
<html lang="ja">
<head>
    <meta charset="UTF-8">
    <title>Document</title>
</head>
<body>
    <h1>タイトル</h1>
    <h2>演習内容</h2>
    <ol id="step1", class="study-list">
        <li>Python基礎</li>
        <li>HTML基礎</li>
    </ol>
    <ol id="step2" class="study-list">
        <li>JS基礎</li>
        <li>Pythonライブラリの基礎</li>
    </ol>
</body>
</html>
"""

soup = BeautifulSoup(html, 'lxml')
print(soup.h1.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.anaconda.com/'
r = requests.get(url)

soup = BeautifulSoup(r.content, 'lxml')
#print(soup.h1.text)
print(soup.find('h2', class_='text-center').text)
#print(soup.find_all('h2', class_='mb-1'))

for h2 in soup.find_all('h2', class_='mb-1'):
    print(h2.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.python.org/'
r = requests.get(url)

soup = BeautifulSoup(r.content, 'lxml')
tag = soup.find('div', class_='shrubbery')

print(tag.find(text='Latest News').parent)
```

```python
import re

import requests
from bs4 import BeautifulSoup

url = 'https://www.python.org/'
r = requests.get(url)

soup = BeautifulSoup(r.content, 'lxml')

# Pyrhonが入っているすべての要素を取得
for text in soup.find_all('a', href=re.compile('docs.python.org')):
    print(text.parent)
```
