```python
from time import sleep
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

options = webdriver.ChromeOptions()

#options.add_argument('--headless')
options.add_argument('--incognito')

driver = webdriver.Chrome(options=options)
driver.implicitly_wait(10)

driver.get('https://www.mizuhobank.co.jp/retail/takarakuji/check/numbers/backnumber/index.html')
sleep(3)
    
soup = BeautifulSoup(driver.page_source, 'lxml')

# ======================抽選データ抽出（過去分）======================
a_tags = soup.select('table.js-backnumber-b a')

base_url = 'https://www.mizuhobank.co.jp/'

deta = []

for a_tag in a_tags[:3]:
    if(a_tag.get('href') != ''):
        detail_url = base_url + a_tag.get('href')
        
        detail_driver = webdriver.Chrome(options=options)
        detail_driver.implicitly_wait(10)

        detail_driver.get(detail_url)
        sleep(3)
        
        detail_soup = BeautifulSoup(detail_driver.page_source, 'lxml')
        
        numbers = detail_soup.select('div.spTableScroll > table > tbody > tr')
        
        for number in numbers:
            kaibetsu = number.select_one('.bgf7f7f7').text
            tyusen = number.select('td.alnRight')
            detam = {
                '回別' : kaibetsu,
                '当選日' : tyusen[0].text,
                '抽選数字' : tyusen[1].text
            }
            deta.append(detam)
        
        detail_driver.quit()
        

# ======================抽選データ抽出（直近分）======================
a_tags = soup.select('table.typeTK > tbody > tr.js-backnumber-temp-a > td.alnRight > a')

for i in range(0, len(a_tags), 2):
    detail_url = base_url + a_tags[i].get('href')
    
    detail_driver = webdriver.Chrome(options=options)
    detail_driver.implicitly_wait(10)

    detail_driver.get(detail_url)
    sleep(3)
        
    detail_soup = BeautifulSoup(detail_driver.page_source, 'lxml')
    
    numbers = detail_soup.select('section.section > div.inner > div > table.typeTK')
    
    for number in numbers:
        kaibetsu = number.select_one('thead > tr > th.js-lottery-issue-pc').text
        tyusenbi = number.select_one('tbody > tr > td.js-lottery-date-pc').text
        tyusen_number = number.select_one('tbody > tr > td.extension > strong.js-lottery-number-pc').text
        detam = {
                '回別' : kaibetsu,
                '当選日' : tyusenbi,
                '抽選数字' : tyusen_number
        }
        deta.append(detam)
    detail_driver.quit()

driver.quit()

df = pd.DataFrame(deta)
print(df)

df.to_csv('mizuho.csv', index=None, encoding='utf-8-sig')
```
