```python
from time import sleep
from selenium import webdriver
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()

#options.add_argument('--headless')
options.add_argument('--incognito')

driver = webdriver.Chrome(options=options)
driver.implicitly_wait(10)

driver.get('https://news.yahoo.co.jp/')
sleep(3)

search_box = driver.find_element_by_css_selector('input.sc-kgoBCf')
sleep(3)

search_box.send_keys('機械学習')
sleep(3)

search_box.submit()

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    sleep(2)

    button = driver.find_elements_by_css_selector('span.button-height-large > button')

    if button:
        button[0].click()
    else:
        break
    sleep(2)

soup = BeautifulSoup(driver.page_source, 'lxml')
a_tags = soup.select('a.newsFeed_item_link')
#a_tags = driver.find_elements_by_css_selector('a.newsFeed_item_link')
#print(news_list)
for i, a_tag in enumerate(a_tags):
    print('='*30, i, '='*30)
    print(a_tag.select_one('div.newsFeed_item_title').text)
    print(a_tag.get('href'))



driver.quit()
#driver.close()
```
